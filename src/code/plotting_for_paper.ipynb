```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
import pandas as pd
data = pd.read_csv('../results/results_summary.csv')
current_13 = pd.read_csv('../results/results_summary_demand=13_posts=original_allocation=current_year=2019.csv')
```

```python
data
```

```python
colours = ['forestgreen', 'gold', 'darkorange', 'red']
fig, ax = plt.subplots(1)
for demand_level, colour in zip([13, 19, 34, 45], colours):
    ys = data[
        (data['Demand Level']==demand_level) & (data['Scenario']=='single')
    ].groupby('Resource Level')['Ambulance Utilisation'].mean() * 100
    ax.plot(range(60, 125, 2), ys, color=colour, label=f"D{demand_level}")
ax.set_ylim(0, 100)
ax.set_ylabel('Ambulance Utilisation (%)', fontsize=16)
ax.set_xlabel('Resource Level', fontsize=16)
ax.set_xlim(60, 124)
ax.set_xticks(range(60, 125, 4))
ax.tick_params(axis='y', which='major', labelsize=14)
ax.set_title('Scenario: Single Vehicle', fontsize=16, y=1.2)
ax.legend(ncol=4, frameon=True, fontsize=13, title="Demand Scenario", bbox_to_anchor=(0.98, 1.23))
plt.subplots_adjust(top=0.75)
fig.savefig('../../../upgraded-octo-barnacle/tex/img/plots/single_AmbulanceUtilisation.pdf')
```
